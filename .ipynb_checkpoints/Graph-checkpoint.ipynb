```python
df_institutions = []
count = 0
for asset in assets:
    count += 1
    ticker = yf.Ticker(asset)
    print(str(count) + '/' + str(len(assets)))
    if ticker.institutional_holders is not None:
        df = ticker.institutional_holders
        if 'Shares' in df.columns:
            df['Ticker'] = asset
            df['%Shares'] = np.floor(df['Shares']/df['Shares'].sum()*100)
            df_institutions.append(df)

dfI = pd.concat(df_institutions)
dfI.to_csv('dataframe.csv')
```

```python
dfI = dfI.drop(columns=['% Out', 'Date Reported', 'Value'])
dfI = dfI.reset_index(drop=True)
```

```python
listInstitutions = list(dfI['Holder'].unique())
listInstitutions
```

```python
dfI
```

```python
holders = dfI.drop(columns=['Shares'])
holders = holders.groupby(['Holder']).mean()
holders['%size'] = holders['%Shares']*100/holders['%Shares'].max()
holders
```

```python
df_nodes = pd.DataFrame(assets)
df_nodes['Type'] = "Ticker"
df_nodes['Size'] = 5
df_nodes = df_nodes.set_index(df_nodes[0])

df_i = holders.copy()
df_i['Type'] = "Holder"
df_i['Size'] = holders['%size']
df_i = df_i.drop(columns=['%Shares', '%size'])
df_nodes = pd.concat([df_nodes, df_i])
df_nodes['Type'] = pd.Categorical(df_nodes['Type'])
df_nodes = df_nodes.drop(columns=[0])
df_nodes
```

```python
list(df_nodes.index)
```

```python
# Creating the graph
g = nx.Graph()
g.add_nodes_from(list(df_nodes.index))
for index, row in dfI.iterrows():
    #print(row['Holder'], row['Ticker'], row['%Shares'])
    g.add_edge(row['Ticker'], row['Holder'], weight=row['%Shares'])
#    g.add_weighted_edges_from(row['Holder'], row['Ticker'])
```

```python
g.nodes()
```

```python

for edge in g.edges():
    exist = False
    print(edge)
    for asset in assets:
        if asset == edge[0]:
            exist = True
    if not exist:
        g.remove_node(asset)
```

```python
# Ploting the graph
plt.figure(figsize=(100,100))
cmap = matplotlib.colors.ListedColormap(['darkorange', 'lightgray', 'dodgerblue'])

nx.draw(g, with_labels=True, node_color=df_nodes['Type'].cat.codes, cmap=cmap, 
        node_size=df_nodes['Size']*10, font_size=11)
plt.show()
plt.savefig('dataframe.svg', format='svg')
```
